## Условие задачи:

Создайте класс LinkedList, реализующий односвязный список. Список должен иметь следующие методы:

__init__(): Инициализирует пустой список. 
__str__(): Преобразует список в строковое представление в виде "| value1, value2, ..., valueN |".
append(obj): Добавляет новый элемент в конец списка.

Пример использования:

1. Создайте объект класса LinkedList.
2. Добавьте несколько элементов в список с помощью метода append.
3. Выведите список на экран.

Ожидаемый результат:

||  # Пустой список
| 22, 2, 3 | # После добавления элементов
Дополнительно:

Node: Создайте класс Node, который представляет собой отдельный элемент списка. Каждый узел должен содержать значение (val) и ссылку на следующий узел (next_node).

Замечания:

В представлении списка разделитель между значениями - запятая, а в начале и конце - вертикальные черты.
В начальном состоянии список пуст.

Помните: Данный код реализует односвязный список.
